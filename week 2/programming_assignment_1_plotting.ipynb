```python
import matplotlib.pyplot as plt
File = open("ex1data1.txt","r")

x = []
y = []

for item in File:
    x.append(float(item.split(",")[0]))
    item = item.strip()
    y.append(float(item.split(",")[1]))

plt.plot(x,y,'rx')
    
plt.xlabel("Population")
plt.ylabel("Profit in 10 thousands")
plt.show()

```
